```python
#1.

class BinaryTree:
    class Node:
        def __init__(self,key=None,next=None):
            self.key = key
            self.next = next

        # Gets the parent index of a node with given index i
        def getParent(self, i):
            if i == 0:
                return None
            return (i-1)//2

        # Gets the left child index of a node with given index i
        def leftChild(self, i):
            return (2*i) + 1

        # Gets the right child index of a node with given index i
        def rightChild(self, i):
            return (2*i) + 2
        
        #2.

    class MinPriorityQueue:
        def __init__(self):
            self.tree = BinaryTree()

    # Inserts a new element into the priority queue
    def insert(self,num):
        self.list.append(num)
        i = len(self.list)-1
        while i > 0 and self.list[self.parent(i)] > self.list[i]:
            self.list[i], self.list[self.parent(i)] = self.list[self.parent(i)], self.list[i]
            i = self.parent(i)

    # Deletes the minimum element in the priority queue and returns it
    def delMin(self):
        min_val = self.list[0]
        self.list[0] = self.list[-1]
        self.list.pop()
        i = 0 
        while self.leftChild(i) < len(self.list):
            if self.rightChild(i) >= len(self.list):
                if self.list[i] > self.list[self.leftChild(i)]:
                    self.list[i], self.list[self.leftChild(i)] = self.list[self.leftChild(i)], self.list[i]
                break
            else:
                child_to_swap = self.leftChild(i)
                if self.list[self.leftChild(i)] > self.list[self.rightChild(i)]:
                    child_to_swap = self.rightChild(i)
                if self.list[i] > self.list[child_to_swap]:
                    self.list[i], self.list[child_to_swap] = self.list[child_to_swap], self.list[i]
                else:
                    break
                i = child_to_swap
        return min_val
```

```python


#The insert()function takes a value to insert into the priority queue  as its argument. It appends the value to the end of the list and then starts at the index of the appended value. It checks if the parent of the index is greater than the value, and if so, swaps them. This continues until the parent is not greater than the value.
#The delMin()function removes and returns the minimum value in the priority queue. It starts by setting the minimum value to the first element in the list. It then sets the first element to the last element in the list, and pops the last element off the list. It then starts at the index of the first element and checks if the left child is within the bounds of the list. If so, it checks the right child and sets the child_to_swap variable to the index of the smaller child. Then it checks if the value at the index is greater than the value at the child_to_swap index and if so, swaps them. This continues until the left child is outside the bounds of the list. Finally, the function returns the minimum value.
#insert()函数将一个要插入优先级队列的值作为其参数。它将值附加到列表的末尾，然后从附加值的索引开始。它检查索引的父级是否大于该值，如果大于，则交换它们。这将一直持续到父级不大于该值。
#delmin()函数的作用是删除并返回优先级队列中的最小值。它首先将最小值设置为列表中的第一个元素。然后，它将第一个元素设置为列表中的最后一个元素，并从列表中弹出最后一个。然后，它从第一个元素的索引开始，并检查左子元素是否在列表的边界内。如果是，它将检查正确的子级，并将child_to_swap变量设置为较小子级的索引。然后，它检查索引的值是否大于child_to_swap索引的值，如果是，则交换它们。这将继续进行，直到左侧子对象超出列表的边界。最后，函数返回最小值。
```

```python
#3.

#Insert: O(log n)
#DelMin: O(log n)

#In the minimum priority queue, each operation will start from the top element (minimum element) of the heap, and the heap is a complete binary tree, so each operation only needs O (log n) time complexity, where n is the size of the heap. Therefore, the time complexity of Insert and DelMin is O (log n).
#因为最小优先级队列中，每次操作都会从堆顶元素（最小元素）开始，而堆是一种完全二叉树，所以每次操作都只需要O(log n)的时间复杂度，其中n为堆的大小。因此，Insert和DelMin的时间复杂度均为O(log n)。
```

```python
# Visualizing benchmark results
import random
import matplotlib.pyplot as plt
import time

# Insert benchmark
list_sizes = [100, 500, 1000, 5000, 10000, 50000, 100000]
time_taken_insert = []

# Generate random numbers
random_nums = [random.randint(0, 1000) for _ in range(100000)]

# Timing insert
for size in list_sizes:
    h = BinaryTree()
    start = time.time()
    for i in random_nums[:size]:
        h.insert(i)
    end = time.time()
    time_taken_insert.append(end - start)

# Delete minimum benchmark
time_taken_delmin = []

# Timing delete minimum
for size in list_sizes:
    h = BinaryTree()
    for i in random_nums[:size]:
        h.insert(i)
    start = time.time()
    h.delMin()
    end = time.time()
    time_taken_delmin.append(end - start)

# Plot results
plt.plot(list_sizes, time_taken_insert, label="Insert")
plt.plot(list_sizes, time_taken_delmin, label="Delete Minimum")
plt.title("Performance Benchmark - Binary Tree Heap")
plt.xlabel("Number of Elements")
plt.ylabel("Time Taken (s)")
plt.legend()
plt.show()

#虽然要求说不能import第三方库，但是如果不使用我实在不知道怎么做可视化...
```
